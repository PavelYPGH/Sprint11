–û–±—Å—É–¥–∏–º –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ –∏—Ö –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ. –ü—Ä–µ–∂–¥–µ –≤—Å–µ–≥–æ, –∫–æ–¥–∏—Ä–æ–≤–∞–Ω–∏–µ –Ω–µ–æ–±—Ö–æ–¥–∏–º–æ, —Ç–∞–∫ –∫–∞–∫ –º–æ–¥–µ–ª–∏ —Ä–∞–±–æ—Ç–∞—é—Ç —Å —á–∏—Å–ª–∞–º–∏, –∞ –Ω–µ —Å–ª–æ–≤–∞–º–∏.

- `last_coffee_type` - 3 –∑–Ω–∞—á–µ–Ω–∏—è.

- `milk_preference` - 6 –∑–Ω–∞—á–µ–Ω–∏–π.
- `coffee_bean_origin` - 6 –∑–Ω–∞—á–µ–Ω–∏–π.
- `subscription_status` - 4 –∑–Ω–∞—á–µ–Ω–∏—è.

–¢–∞–∫ –∫–∞–∫ –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –∑–Ω–∞—á–µ–Ω–∏–π –≤ –Ω–∏—Ö –Ω–µ–≤–µ–ª–∏–∫–æ, –¥–ª—è –∫–æ–¥–∏—Ä–æ–≤–∫–∏ –±—É–¥–µ—Ç –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å—Å—è OneHotEncoder.

–¢–µ–ø–µ—Ä—å —Ä–∞—Å—Å–º–æ—Ç—Ä–∏–º –≤—ã–±—Ä–æ—Å—ã —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —Å—Ç–æ–ª–±—Ü—ã —Å–ª–µ–¥—É—é—â–∏–µ:

- `days_since_last_order`

- `order_frequency_week`
- `avg_order_value`
- `median_order_value`
- `total_spent_last_month`
- `total_spent_last_week`
- `discount_usage_rate`
- `seasonal_menu_tried`

- `app_crashes_last_month`

```python
numeric_cols = [
    'days_since_last_order',
    'order_frequency_month', 
    'order_frequency_week',
    'avg_order_value',
    'median_order_value',
    'total_spent_last_month',
    'total_spent_last_week',
    'discount_usage_rate',
    'seasonal_menu_tried',
    'app_opens_per_week',
    'review_rating_last_10',
    'app_crashes_last_month',
    'days_since_last_promo'
]

for col in numeric_cols:
    print(f"{'='*50}")
    print(f"{col.upper():^50}")
    print(f"{'='*50}")
    
    stats = df[col].describe()
    print(stats.round(2))
    
    q99 = df[col].quantile(0.99)
    q01 = df[col].quantile(0.01)
    
    print(f"\n–î–æ–ø–æ–ª–Ω–∏—Ç–µ–ª—å–Ω–æ:")
    print(f"99% –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å: {q99:.2f}")
    print(f"1% –ø–µ—Ä—Ü–µ–Ω—Ç–∏–ª—å: {q01:.2f}")
    print(f"–†–∞–∑–º–∞—Ö (max-min): {df[col].max() - df[col].min():.2f}")
    
    outliers_99 = df[df[col] > q99]
    print(f"–ó–Ω–∞—á–µ–Ω–∏–π >99%: {len(outliers_99)} ({(len(outliers_99)/len(df)*100):.1f}%)")
    
    median = df[col].median()
    mean = df[col].mean()
    
    print()
```

–ù–∞–º –±—É–¥–µ—Ç –≤—ã–≥–æ–¥–Ω–æ –æ—Å—Ç–∞–≤–∏—Ç—å –¥–∞–Ω–Ω—ã–µ –∫–∞–∫ –µ—Å—Ç—å. –ù–æ –≤–∞–∂–Ω–æ –±—É–¥–µ—Ç –ø–æ—Å–º–æ—Ç—Ä–µ—Ç—å –Ω–∞ –≥—Ä–∞—Ñ–∏–∫–∏ - –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã –¥–ª—è –Ω–∞—à–∏—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤, —É–≤–∏–¥–µ—Ç—å —Ö–∞—Ä–∞–∫—Ç–µ—Ä —Ä–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏—è, –ø–æ–Ω—è—Ç—å, –∫–∞–∫ –º–æ–∂–µ–º ¬´–ø–æ–º–æ—á—å¬ª –º–æ–¥–µ–ª—å–∫–∞–º —É—Å–≤–æ–∏—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏. –ù–∞–ø—Ä–∏–º–µ—Ä, –ø—Ä–∏–º–µ–Ω–∏–≤ –∫ –Ω–∏–º –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—Ä–æ–≤–∞–Ω–∏–µ –∏–ª–∏ –≤–∑—è—Ç—å –∫–æ—Ä–µ–Ω—å –∏–∑ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –¥–ª—è —Ç–æ–≥–æ, —á—Ç–æ–±—ã —Å–≥–ª–∞–¥–∏—Ç—å –≤—ã–±—Ä–æ—Å—ã. –õ–∏–±–æ –Ω–∞–æ–±–æ—Ä–æ—Ç –¥–ª—è –Ω–µ–∫–æ—Ç–æ—Ä—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ –ª–æ–≥–∏–∫–µ –∏ —Å–º—ã—Å–ª—É —Å–¥–µ–ª–∞—Ç—å –≤–æ–∑–≤–µ–¥–µ–Ω–∏–µ –≤ –∫–≤–∞–¥—Ä–∞—Ç (–ø—Ä–µ–¥–ø–æ–ª–æ–∂—É, —á—Ç–æ –¥–ª—è –æ—à–∏–±–æ–∫ —Å–µ—Ä–≤–∏—Å–∞ —Ç–∞–∫–æ–µ –º–æ–∂–µ—Ç —Å—Ä–∞–±–æ—Ç–∞—Ç—å). 

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()
corr_matrix = df[numeric_cols].corr()

plt.figure(figsize=(16, 12))
sns.heatmap(corr_matrix, 
            annot=True, 
            fmt=".2f", 
            cmap='coolwarm',
            center=0,
            square=True,
            linewidths=0.5,
            cbar_kws={"shrink": 0.8})
plt.title('–ú–∞—Ç—Ä–∏—Ü–∞ –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–π (—á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏)', fontsize=16)
plt.tight_layout()
plt.show()

plt.figure(figsize=(10, 8))
churn_corr = corr_matrix['churn'].drop('churn').sort_values()
bars = plt.barh(churn_corr.index, churn_corr.values, 
                color=np.where(churn_corr.values > 0, 'red', 'blue'))

plt.title('–ö–æ—Ä—Ä–µ–ª—è—Ü–∏—è –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ —Å churn', fontsize=14)
plt.xlabel('–ö–æ—ç—Ñ—Ñ–∏—Ü–∏–µ–Ω—Ç –∫–æ—Ä—Ä–µ–ª—è—Ü–∏–∏')
plt.grid(axis='x', alpha=0.3, linestyle='--')
plt.axvline(x=0, color='black', linewidth=0.8)

for bar in bars:
    width = bar.get_width()
    plt.text(width + (0.01 if width >= 0 else -0.01), 
             bar.get_y() + bar.get_height()/2,
             f'{width:.3f}', 
             va='center',
             ha='left' if width >= 0 else 'right',
             fontsize=9)

plt.tight_layout()
plt.show()
```

- –ù–µ –∫–æ—Ä—Ä–µ–ª–∏—Ä—É—é—â–∏–µ –Ω–∏ —Å —á–µ–º –ø—Ä–∏–∑–Ω–∞–∫–∏, –∫—Ä–æ–º–µ —Å–∞–º–∏—Ö —Å–µ–±—è, –±—ã–ª–∏ —É–¥–∞–ª–µ–Ω—ã.

–†–µ–∫–æ–º–µ–Ω–¥—É—é –≤–µ—Ä–Ω—É—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏ –ø–æ –º–∞–∫—Å–∏–º—É–º—É. 

## –≠—Ç–∞–ø 3. –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∞ –¥–∞–Ω–Ω—ã—Ö

1. –†–∞–∑–¥–µ–ª–∏—Ç–µ –¥–∞–Ω–Ω—ã–µ –≤ –ø—Ä–æ–ø–æ—Ä—Ü–∏–∏ 80 –∫ 20. 20% –¥–∞–Ω–Ω—ã—Ö –æ—Ç–ª–æ–∂–∏—Ç–µ –¥–ª—è —Ç–µ—Å—Ç–∞. –û—Å—Ç–∞–ª—å–Ω—ã–µ –∏—Å–ø–æ–ª—å–∑—É–π—Ç–µ –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –∏ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –º–æ–¥–µ–ª–∏.

2. –ü—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∞–π—Ç–µ –¥–∞–Ω–Ω—ã–µ. –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∏–Ω—Ñ–æ—Ä–º–∞—Ü–∏—é –æ –ø—Ä–æ–ø—É—Å–∫–∞—Ö –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–∞—Ö —Ç–æ–ª—å–∫–æ –∏–∑ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–∏.

   - –°–æ–∑–¥–∞–π—Ç–µ –ø–∞–π–ø–ª–∞–π–Ω, –∫–æ—Ç–æ—Ä—ã–π –æ–±—Ä–∞–±–æ—Ç–∞–µ—Ç –ø—Ä–æ–ø—É—Å–∫–∏ –∏ –≤—ã–±—Ä–æ—Å—ã.

   - –°–æ–∑–¥–∞–π—Ç–µ –ø–∞–π–ø–ª–∞–π–Ω, –∫–æ—Ç–æ—Ä—ã–π –æ–±—Ä–∞–±–æ—Ç–∞–µ—Ç –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.

   - –°–æ–∑–¥–∞–π—Ç–µ –ø–∞–π–ø–ª–∞–π–Ω, –∫–æ—Ç–æ—Ä—ã–π –æ–±—Ä–∞–±–æ—Ç–∞–µ—Ç —á–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏: –ø—Ä–æ–≤–µ–¥—ë—Ç –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ –∏ –Ω–æ—Ä–º–∞–ª–∏–∑–∞—Ü–∏—é.



–°–Ω–∞—á–∞–ª–∞ —Ä–∞–∑–¥–µ–ª–∏–º –¥–∞–Ω–Ω—ã–µ –∏ –ø—Ä–∏–º–µ–Ω–∏–º —Ñ–∏–∫—Å–∏—Ä–æ–≤–∞–Ω–Ω—ã–π –≤ –Ω–∞—á–∞–ª–µ –ø—Ä–æ–µ–∫—Ç–∞ RANDOM_STATE.

```python
X = df.drop('churn', axis=1)
y = df['churn']

X_train, X_test, y_train, y_test = train_test_split(
    X, y, 
    test_size=0.2,
    random_state=RANDOM_STATE,
    stratify=y
)

print(f"–í—Å–µ–≥–æ –¥–∞–Ω–Ω—ã—Ö: {len(df)}")
print(f"–û–±—É—á–∞—é—â–∞—è –≤—ã–±–æ—Ä–∫–∞ (80%): {len(X_train)} —Å—Ç—Ä–æ–∫")
print(f"–¢–µ—Å—Ç–æ–≤–∞—è –≤—ã–±–æ—Ä–∫–∞ (20%): {len(X_test)} —Å—Ç—Ä–æ–∫")
print(f"\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ churn –≤ –æ–±—É—á–∞—é—â–µ–π –≤—ã–±–æ—Ä–∫–µ:")
print(y_train.value_counts(normalize=True).round(3))
print(f"\n–†–∞—Å–ø—Ä–µ–¥–µ–ª–µ–Ω–∏–µ churn –≤ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:")
print(y_test.value_counts(normalize=True).round(3))
```

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞  <a class="tocSkip"> </h2>

üëç –û—Ç–ª–∏—á–Ω–æ, —á—Ç–æ –∏—Å–ø–æ–ª—å–∑—É–µ—à—å —Å—Ç—Ä–∞—Ç–∏—Ñ–∏–∫–∞—Ü–∏—é.

–¢–∞–∫–∂–µ –≤—ã–¥–µ–ª–∏–º —á–∏—Å–ª–æ–≤—ã–µ –∏ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.

```python
numeric_cols = [
    'order_frequency_month',
    'order_frequency_week', 
    'avg_order_value',
    'median_order_value',
    'total_spent_last_month',
    'total_spent_last_week',
    'app_opens_per_week',
    'app_crashes_last_month'
]

categorical_cols = [
    'last_coffee_type',
    'preferred_roast',
    'milk_preference',
    'coffee_bean_origin',
    'subscription_status'
]
```

–ò —Å–¥–µ–ª–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é, —Ñ–∏–ª—å—Ç—Ä—É—é—â—É—é —á–∏—Å–ª–æ–≤—ã–µ –∑–Ω–∞—á–µ–Ω–∏—è > 0 –∏ –ø–æ 99 –ø—Ä–æ—Ü–µ–Ω—Ç–∏–ª—é.

<div class="alert alert-warning">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>
    
<b>–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏üí°:</b> 
        
–ï—Å–ª–∏ –ø—Ä–∏–∑–Ω–∞–∫–∏ —Ä–µ–∞–ª–∏—Å—Ç–∏—á–Ω—ã, —Ä–µ–∫–æ–º–µ–Ω–¥—É—é —Ç–∞–∫ –Ω–µ –¥–µ–ª–∞—Ç—å. –í—ã—à–µ –æ–±—è–∑–∞—Ç–µ–ª—å–Ω–æ –Ω—É–∂–Ω–æ –±—É–¥–µ—Ç –≤–∏–∑—É–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –ø—Ä–∏–∑–Ω–∞–∫–∏, –ø—Ä–æ–∫–æ–º–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å –≥–∏—Å—Ç–æ–≥—Ä–∞–º–º—ã.

```python
def cap_outliers_99(X):

    X_capped = X.copy()
    
    for col_idx in range(X.shape[1]):
        col_data = X[:, col_idx]
        
        cap_value = np.percentile(col_data, 99)
        
        mask = (col_data > cap_value) & (col_data > 0)
        col_data[mask] = cap_value
        
        X_capped[:, col_idx] = col_data
    
    return X_capped
```

–ù–∞—á–Ω—ë–º —Å –ø–∞–π–ø–ª–∞–π–Ω–∞ –¥–ª—è —á–∏—Å–ª–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –ó–∞–ø–æ–ª–Ω—è–µ–º –ø—Ä–æ–ø—É—Å–∫–∏ –º–µ–¥–∏–∞–Ω–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏, —Ñ–∏–ª—å—Ç—Ä—É–µ–º –∏ —Å—Ç–∞–Ω–¥–∞—Ä—Ç–∏–∑–∏—Ä—É–µ–º.

```python
print("–ü–∞–π–ø–ª–∞–π–Ω: –û–±—Ä–∞–±–æ—Ç–∫–∞ –ø—Ä–æ–ø—É—Å–∫–æ–≤ –∏ –≤—ã–±—Ä–æ—Å–æ–≤ + –º–∞—Å—à—Ç–∞–±–∏—Ä–æ–≤–∞–Ω–∏–µ")
print("-" * 40)
numeric_processing_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('outlier_capper', FunctionTransformer(cap_outliers_99, validate=False)),
    ('scaler', StandardScaler())
])
print("–°–æ–∑–¥–∞–Ω: numeric_processing_pipeline")
```

–¢–µ–ø–µ—Ä—å –ø–∞–π–ø–ª–∞–π–Ω –¥–ª—è –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –ó–∞–ø–æ–ª–Ω—è–µ–º —Å–∞–º—ã–º–∏ –ø–æ–ø—É–ª—è—Ä–Ω—ã–º–∏ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –∏ –∫–æ–¥–∏—Ä—É–µ–º.

```python
print("–ü–∞–π–ø–ª–∞–π–Ω: –û–±—Ä–∞–±–æ—Ç–∫–∞ –∫–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
print("-" * 40)
categorical_processing_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(sparse = False, handle_unknown='ignore'))
])
print("–°–æ–∑–¥–∞–Ω: categorical_processing_pipeline")
```

<div class="alert alert-warning">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞ <a class="tocSkip"> </h2>
    
<b>–ù–µ–∫–æ—Ç–æ—Ä—ã–µ –∑–∞–º–µ—á–∞–Ω–∏—è –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏üí°:</b> 
        
–ß—Ç–æ–±—ã –∏–∑–±–µ–≥–∞—Ç—å –¥–∞–º–º–∏-–ª–æ–≤—É—à–∫–∏ —Å—Ç–æ–∏—Ç –ø—Ä–æ–ø–∏—Å–∞—Ç—å –≤ OneHotEncoder() –ø–∞—Ä–∞–º–µ—Ç—Ä drop='first'
        
–ü—Ä–∞–∫—Ç–∏–∫—É–º –ø–æ–¥–≥–æ—Ç–æ–≤–∏–ª –∏–Ω—Å—Ç—Ä—É–∫—Ü–∏—é, –∫–∞–∫ –∏—Å–ø–æ–ª—å–∑–æ–≤–∞—Ç—å OHE: https://colab.research.google.com/drive/1_gAMXcQKoCShB_l8FNtYEejMnosm9mvt?usp=sharing 




–ò –¥–µ–ª–∞–µ–º —Ç—Ä–∞–Ω—Å—Ñ–æ—Ä–º–µ—Ä.

```python
print("\n" + "=" * 60)
print("–û–ë–™–ï–î–ò–ù–ï–ù–ò–ï –í –ï–î–ò–ù–´–ô –ü–†–ï–ü–†–û–¶–ï–°–°–û–†")
print("=" * 60)

full_preprocessor = ColumnTransformer(
    transformers=[
        ('numeric_processing', numeric_processing_pipeline, numeric_cols),
        ('categorical_processing', categorical_processing_pipeline, categorical_cols)
    ]
)

print("–°–æ–∑–¥–∞–Ω full_preprocessor:")
print(full_preprocessor)
```

–û–±—É—á–∞–µ–º –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–æ—Ä –Ω–∞ —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö –∏ –ø—Ä–∏–º–µ–Ω—è–µ–º –∫ –Ω–∏–º –∂–µ –∏ –∫ —Ç–µ—Å—Ç–æ–≤—ã–º –¥–∞–Ω–Ω—ã–º.

```python
print("\n" + "=" * 60)
print("–ü–†–ò–ú–ï–ù–ï–ù–ò–ï –ü–†–ï–ü–†–û–¶–ï–°–°–û–†–ê –ö –î–ê–ù–ù–´–ú")
print("=" * 60)

X_train_processed = full_preprocessor.fit_transform(X_train)

X_test_processed = full_preprocessor.transform(X_test)

print(f"–†–∞–∑–º–µ—Ä—ã –ø–æ—Å–ª–µ –æ–±—Ä–∞–±–æ—Ç–∫–∏:")
print(f"X_train: {X_train_processed.shape}")
print(f"X_test:  {X_test_processed.shape}")
```

–ü—Ä–æ–≤–µ—Ä–∏–º –∫–∞–∫ –≤—Å—ë –∑–∞–ø–æ–ª–Ω–∏–ª–æ—Å—å –∏ –∑–∞–∫–æ–¥–∏—Ä–æ–≤–∞–ª–æ—Å—å.

## –≠—Ç–∞–ø 4. –û–±—É—á–µ–Ω–∏–µ –º–æ–¥–µ–ª–∏

1. –û–±—É—á–∏—Ç–µ –±–∞–∑–æ–≤—É—é –≤–µ—Ä—Å–∏—é –º–æ–¥–µ–ª–∏.
   - –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –¥–ª—è —ç—Ç–æ–≥–æ –ø—Ä–æ—Å—Ç—ã–µ —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∏–µ –º–æ–¥–µ–ª–∏.

   - –ò—Å–ø–æ–ª—å–∑—É–π—Ç–µ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –¥–ª—è –æ–±—É—á–µ–Ω–∏—è –º–æ–¥–µ–ª–∏.

2. –ü–æ—Å—á–∏—Ç–∞–π—Ç–µ –º–µ—Ç—Ä–∏–∫–∏, –ø–æ—Å—Ç–∞–≤–ª–µ–Ω–Ω—ã–µ –≤ –∑–∞–¥–∞—á–µ. –û–ø–∏—Ä–∞—è—Å—å –Ω–∞ –Ω–∏—Ö, —Å–¥–µ–ª–∞–π—Ç–µ –≤—ã–≤–æ–¥ –æ –∫–∞—á–µ—Å—Ç–≤–µ –º–æ–¥–µ–ª–∏.

–°–æ–∑–¥–∞–¥–∏–º –º–æ–¥–µ–ª–∏ DummyClassifier –∏ LogisticRegression.

```python
dummy_model = DummyClassifier(
    strategy='stratified',
    random_state = RANDOM_STATE
)

logreg_model = LogisticRegression(
    random_state = RANDOM_STATE,
    class_weight='balanced',
    max_iter=1000,
    solver='liblinear'
)

print("–ú–æ–¥–µ–ª–∏ —Å–æ–∑–¥–∞–Ω—ã")
```

–ü—Ä–æ–≤–µ–¥—ë–º –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é —Å 5-—é —Ñ–æ–ª–¥–∞–º–∏.

```python
cv = StratifiedKFold(
    n_splits=5,
    shuffle=True,
    random_state = RANDOM_STATE
)

print("–í—ã–ø–æ–ª–Ω—è–µ–º –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –¥–ª—è LogisticRegression")

y_pred_proba_cv = cross_val_predict(
    logreg_model, 
    X_train_processed, 
    y_train,
    cv=cv,
    method='predict_proba'
)[:, 1]

y_pred_cv = cross_val_predict(
    logreg_model,
    X_train_processed,
    y_train,
    cv=cv,
    method='predict'
)

print("–ö—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è –∑–∞–≤–µ—Ä—à–µ–Ω–∞")
print(f"   –ü–æ–ª—É—á–µ–Ω–æ {len(y_pred_proba_cv)} –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π –≤–µ—Ä–æ—è—Ç–Ω–æ—Å—Ç–µ–π")

precision_cv, recall_cv, _ = precision_recall_curve(y_train, y_pred_proba_cv)
pr_auc_cv = auc(recall_cv, precision_cv)

print(f"PR-AUC: {pr_auc_cv:.4f}")
```

–î–∞–≤–∞–π –≤—ã—à–µ –ø–æ–ø—Ä–æ–±—É–µ–º –±–æ–ª–µ–µ –ø–æ–¥—Ä–æ–±–Ω–æ –ø—Ä–æ–≤–µ—Å—Ç–∏ –∞–Ω–∞–ª–∏–∑ –¥–∞–Ω–Ω—ã—Ö, –ø–æ–Ω—è—Ç—å –∏—Ö –ø—Ä–∏—Ä–æ–¥—É –∏ –≤–æ–∑–º–æ–∂–Ω–æ—Å—Ç–∏ –ø–æ —É–ª—É—á—à–µ–Ω–∏—é –¥–ª—è –º–æ–¥–µ–ª–µ–π, –∞ –∑–¥–µ—Å—å —É–∂–µ –æ–ø–∏—Ä–∞—è—Å—å –Ω–∞ —ç—Ç—É –∞—Ä–≥—É–º–µ–Ω—Ç–∞—Ü–∏—é –±—É–¥–µ–º –≤–æ–∑–¥–∞–≤–∞—Ç—å –Ω–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏.

```python
X_train_extended = X_train.copy()
X_test_extended = X_test.copy()

print("–ò—Å—Ö–æ–¥–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏:", list(X_train.columns))

print("\n–ü–†–ò–ó–ù–ê–ö–ò –ü–û–í–ï–î–ï–ù–ò–Ø:")
# –ê–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∫–ª–∏–µ–Ω—Ç–∞: –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –Ω–µ–¥–µ–ª—å–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã –∫ –º–µ—Å—è—á–Ω–æ–π
X_train_extended['activity_ratio'] = X_train_extended['order_frequency_week'] / (X_train_extended['order_frequency_month'] + 1e-6)
X_test_extended['activity_ratio'] = X_test_extended['order_frequency_week'] / (X_test_extended['order_frequency_month'] + 1e-6)
print("‚Ä¢ activity_ratio - –æ—Ç–Ω–æ—à–µ–Ω–∏–µ weekly/monthly —á–∞—Å—Ç–æ—Ç—ã –∑–∞–∫–∞–∑–æ–≤")

# –¢–µ–Ω–¥–µ–Ω—Ü–∏—è —Ç—Ä–∞—Ç: –∏–∑–º–µ–Ω–µ–Ω–∏–µ —Ç—Ä–∞—Ç –∑–∞ –ø–æ—Å–ª–µ–¥–Ω—é—é –Ω–µ–¥–µ–ª—é –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –º–µ—Å—è—Ü–∞
X_train_extended['spending_trend'] = X_train_extended['total_spent_last_week'] / (X_train_extended['total_spent_last_month'] + 1e-6)
X_test_extended['spending_trend'] = X_test_extended['total_spent_last_week'] / (X_test_extended['total_spent_last_month'] + 1e-6)
print("‚Ä¢ spending_trend - –æ—Ç–Ω–æ—à–µ–Ω–∏–µ weekly/monthly —Ç—Ä–∞—Ç")

# –ü—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ —Ü–µ–Ω–æ–≤—ã—Ö —Ö–∞—Ä–∞–∫—Ç–µ—Ä–∏—Å—Ç–∏–∫ (—Å–≥–ª–∞–∂–∏–≤–∞–Ω–∏–µ)
print("\n–ü–†–ò–ó–ù–ê–ö–ò –¶–ï–ù–´ (—Å–≥–ª–∞–∂–µ–Ω–Ω—ã–µ):")
# –°–≥–ª–∞–∂–∏–≤–∞–µ–º –±–æ–ª—å—à–∏–µ –∑–Ω–∞—á–µ–Ω–∏—è —á–µ—Ä–µ–∑ –∫–≤–∞–¥—Ä–∞—Ç–Ω—ã–π –∫–æ—Ä–µ–Ω—å
X_train_extended['sqrt_avg_order'] = np.sqrt(X_train_extended['avg_order_value'] + 1)
X_test_extended['sqrt_avg_order'] = np.sqrt(X_test_extended['avg_order_value'] + 1)
print("‚Ä¢ sqrt_avg_order - –∫–æ—Ä–µ–Ω—å –∏–∑ —Å—Ä–µ–¥–Ω–µ–≥–æ —á–µ–∫–∞")

X_train_extended['sqrt_median_order'] = np.sqrt(X_train_extended['median_order_value'] + 1)
X_test_extended['sqrt_median_order'] = np.sqrt(X_test_extended['median_order_value'] + 1)
print("‚Ä¢ sqrt_median_order - –∫–æ—Ä–µ–Ω—å –∏–∑ –º–µ–¥–∏–∞–Ω–Ω–æ–≥–æ —á–µ–∫–∞")

# –ü—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ —á–∞—Å—Ç–æ—Ç—ã (—É—Å–∏–ª–µ–Ω–∏–µ –≤–ª–∏—è–Ω–∏—è)
print("\n–ü–†–ò–ó–ù–ê–ö–ò –ß–ê–°–¢–û–¢–´ (—É—Å–∏–ª–µ–Ω–Ω—ã–µ):")
# –£—Å–∏–ª–∏–≤–∞–µ–º –≤–ª–∏—è–Ω–∏–µ —á–∞—Å—Ç–æ—Ç—ã –∑–∞–∫–∞–∑–æ–≤ —á–µ—Ä–µ–∑ –∫–≤–∞–¥—Ä–∞—Ç
X_train_extended['squared_freq_week'] = X_train_extended['order_frequency_week'] ** 2
X_test_extended['squared_freq_week'] = X_test_extended['order_frequency_week'] ** 2
print("‚Ä¢ squared_freq_week - –∫–≤–∞–¥—Ä–∞—Ç –Ω–µ–¥–µ–ª—å–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã")

X_train_extended['squared_freq_month'] = X_train_extended['order_frequency_month'] ** 2
X_test_extended['squared_freq_month'] = X_test_extended['order_frequency_month'] ** 2
print("‚Ä¢ squared_freq_month - –∫–≤–∞–¥—Ä–∞—Ç –º–µ—Å—è—á–Ω–æ–π —á–∞—Å—Ç–æ—Ç—ã")

# –ü—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –∫–æ–º–±–∏–Ω–∞—Ü–∏–π
print("\n–ö–û–ú–ë–ò–ù–ò–†–û–í–ê–ù–ù–´–ï –ü–†–ò–ó–ù–ê–ö–ò:")
# –≠—Ñ—Ñ–µ–∫—Ç–∏–≤–Ω–æ—Å—Ç—å –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–Ω–∏—è –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è
X_train_extended['app_efficiency'] = X_train_extended['order_frequency_week'] / (X_train_extended['app_opens_per_week'] + 1)
X_test_extended['app_efficiency'] = X_test_extended['order_frequency_week'] / (X_test_extended['app_opens_per_week'] + 1)
print("‚Ä¢ app_efficiency - –∑–∞–∫–∞–∑—ã –Ω–∞ –æ–¥–Ω–æ –æ—Ç–∫—Ä—ã—Ç–∏–µ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è")

# –°—Ä–µ–¥–Ω—è—è —Å—Ç–æ–∏–º–æ—Å—Ç—å –∑–∞ –æ–¥–Ω–æ –æ—Ç–∫—Ä—ã—Ç–∏–µ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è
X_train_extended['value_per_app_open'] = X_train_extended['total_spent_last_week'] / (X_train_extended['app_opens_per_week'] + 1)
X_test_extended['value_per_app_open'] = X_test_extended['total_spent_last_week'] / (X_test_extended['app_opens_per_week'] + 1)
print("‚Ä¢ value_per_app_open - —Ç—Ä–∞—Ç—ã –Ω–∞ –æ–¥–Ω–æ –æ—Ç–∫—Ä—ã—Ç–∏–µ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è")

# –ë–∏–Ω–∞—Ä–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏ –Ω–∞ –æ—Å–Ω–æ–≤–µ –ª–æ–≥–∏–∫–∏
print("\n–ë–ò–ù–ê–†–ù–´–ï –ü–†–ò–ó–ù–ê–ö–ò:")
# –ö–ª–∏–µ–Ω—Ç —Å –Ω–∏–∑–∫–æ–π –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å—é (–º–µ–Ω—å—à–µ 2 –∑–∞–∫–∞–∑–æ–≤ –≤ –Ω–µ–¥–µ–ª—é)
X_train_extended['low_activity'] = (X_train_extended['order_frequency_week'] < 2).astype(int)
X_test_extended['low_activity'] = (X_test_extended['order_frequency_week'] < 2).astype(int)
print("‚Ä¢ low_activity - –Ω–∏–∑–∫–∞—è –∞–∫—Ç–∏–≤–Ω–æ—Å—Ç—å (<2 –∑–∞–∫–∞–∑–æ–≤ –≤ –Ω–µ–¥–µ–ª—é)")

# –ö–ª–∏–µ–Ω—Ç —Å –ø–∞–¥–∞—é—â–∏–º–∏ —Ç—Ä–∞—Ç–∞–º–∏
X_train_extended['spending_decline'] = (X_train_extended['spending_trend'] < 0.5).astype(int)
X_test_extended['spending_decline'] = (X_test_extended['spending_trend'] < 0.5).astype(int)
print("‚Ä¢ spending_decline - —Ä–µ–∑–∫–æ–µ –ø–∞–¥–µ–Ω–∏–µ —Ç—Ä–∞—Ç (>50% —Å–Ω–∏–∂–µ–Ω–∏–µ)")

# –ü—Ä–æ–±–ª–µ–º–Ω—ã–π –∫–ª–∏–µ–Ω—Ç (–º–Ω–æ–≥–æ —Å–±–æ–µ–≤)
X_train_extended['problematic_user'] = (X_train_extended['app_crashes_last_month'] > 3).astype(int)
X_test_extended['problematic_user'] = (X_test_extended['app_crashes_last_month'] > 3).astype(int)
print("‚Ä¢ problematic_user - –º–Ω–æ–≥–æ —Å–±–æ–µ–≤ –ø—Ä–∏–ª–æ–∂–µ–Ω–∏—è (>3 –≤ –º–µ—Å—è—Ü)")

print(f"\n–°–æ–∑–¥–∞–Ω–æ {len(X_train_extended.columns) - len(X_train.columns)} –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤")
print(f"–í—Å–µ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: {len(X_train_extended.columns)}")
```

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞  <a class="tocSkip"> </h2>

üëç –í—Å–µ —Å—É–ø–µ—Ä, –≤—Å–µ –ª–æ–≥–∏—á–Ω–æ –∏ –æ—Ç–ª–∏—á–Ω–æ.

–¢–µ–ø–µ—Ä—å –Ω—É–∂–Ω–æ –æ–±–Ω–æ–≤–∏—Ç—å –ø–∞–π–ø–ª–∞–π–Ω—ã –∏ –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–æ—Ä.

```python
numeric_cols_extended = numeric_cols + [
    'activity_ratio',
    'spending_trend',
    'sqrt_avg_order',
    'sqrt_median_order',
    'squared_freq_week',
    'squared_freq_month',
    'app_efficiency',
    'value_per_app_open',
    'low_activity',
    'spending_decline',
    'problematic_user'
]

categorical_cols_extended = categorical_cols

print(f"–ß–∏—Å–ª–æ–≤—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏: {len(numeric_cols_extended)}")
print(f"–ö–∞—Ç–µ–≥–æ—Ä–∏–∞–ª—å–Ω—ã–µ –ø—Ä–∏–∑–Ω–∞–∫–∏: {len(categorical_cols_extended)}")
print(f"–í—Å–µ–≥–æ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤ –ø–æ—Å–ª–µ OHE: ‚âà{len(numeric_cols_extended) + len(categorical_cols_extended)*3}")

def cap_outliers_extended(X):
    X_capped = X.copy()
    
    for col_idx in range(X.shape[1]):
        col_data = X[:, col_idx]
        
        cap_value = np.percentile(col_data, 99)
        
        mask = (col_data > cap_value) & (col_data > 0)
        col_data[mask] = cap_value
        
        X_capped[:, col_idx] = col_data
    
    return X_capped

numeric_processing_pipeline_extended = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('outlier_capper', FunctionTransformer(cap_outliers_extended, validate=False)),
    ('scaler', StandardScaler())
])

categorical_processing_pipeline_extended = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OneHotEncoder(sparse=False, handle_unknown='ignore'))
])

full_preprocessor_extended = ColumnTransformer(
    transformers=[
        ('numeric_processing', numeric_processing_pipeline_extended, numeric_cols_extended),
        ('categorical_processing', categorical_processing_pipeline_extended, categorical_cols_extended)
    ]
)

print("–ù–æ–≤—ã–π –ø—Ä–µ–ø—Ä–æ—Ü–µ—Å—Å–æ—Ä —Å–æ–∑–¥–∞–Ω")
```

–ò –ø—Ä–∏–º–µ–Ω–∏–º –µ–≥–æ.

```python
X_train_processed_extended = full_preprocessor_extended.fit_transform(X_train_extended)
X_test_processed_extended = full_preprocessor_extended.transform(X_test_extended)

print(f"–†–∞–∑–º–µ—Ä—ã –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö –¥–∞–Ω–Ω—ã—Ö:")
print(f"X_train (—Å—Ç–∞—Ä—ã–π): {X_train_processed.shape}")
print(f"X_train (–Ω–æ–≤—ã–π):  {X_train_processed_extended.shape}")
print(f"–£–≤–µ–ª–∏—á–µ–Ω–∏–µ –ø—Ä–∏–∑–Ω–∞–∫–æ–≤: +{X_train_processed_extended.shape[1] - X_train_processed.shape[1]}")
```

–ù–∞–∫–æ–Ω–µ—Ü, –æ–±—É—á–∞–µ–º –∏ —Å—Ä–∞–≤–Ω–∏–≤–∞–µ–º –Ω–æ–≤—É—é –º–æ–¥–µ–ª—å.

```python
logreg_model_extended = LogisticRegression(
    random_state=RANDOM_STATE,
    class_weight='balanced',
    max_iter=1000,
    solver='liblinear'
)

cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

print("–ë–ê–ó–û–í–ê–Ø –ú–û–î–ï–õ–¨ (–±–µ–∑ –Ω–æ–≤—ã—Ö –ø—Ä–∏–∑–Ω–∞–∫–æ–≤):")
print("-" * 50)

y_pred_proba_cv_base = cross_val_predict(
    logreg_model,
    X_train_processed,
    y_train,
    cv=cv,
    method='predict_proba'
)[:, 1]

precision_cv_base, recall_cv_base, _ = precision_recall_curve(y_train, y_pred_proba_cv_base)
pr_auc_cv_base = auc(recall_cv_base, precision_cv_base)

y_pred_cv_base = cross_val_predict(
    logreg_model,
    X_train_processed,
    y_train,
    cv=cv,
    method='predict'
)

print(f"   PR-AUC (–∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è): {pr_auc_cv_base:.4f}")
print(f"   Precision: {precision_score(y_train, y_pred_cv_base):.3f}")
print(f"   Recall: {recall_score(y_train, y_pred_cv_base):.3f}")
print(f"   F1-score: {f1_score(y_train, y_pred_cv_base):.3f}")

print("\n–ú–û–î–ï–õ–¨ –° –ù–û–í–´–ú–ò –ü–†–ò–ó–ù–ê–ö–ê–ú–ò:")
print("-" * 50)

y_pred_proba_cv_extended = cross_val_predict(
    logreg_model_extended,
    X_train_processed_extended,
    y_train,
    cv=cv,
    method='predict_proba'
)[:, 1]

precision_cv_extended, recall_cv_extended, _ = precision_recall_curve(y_train, y_pred_proba_cv_extended)
pr_auc_cv_extended = auc(recall_cv_extended, precision_cv_extended)

y_pred_cv_extended = cross_val_predict(
    logreg_model_extended,
    X_train_processed_extended,
    y_train,
    cv=cv,
    method='predict'
)

print(f"   PR-AUC (–∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è): {pr_auc_cv_extended:.4f}")
print(f"   Precision: {precision_score(y_train, y_pred_cv_extended):.3f}")
print(f"   Recall: {recall_score(y_train, y_pred_cv_extended):.3f}")
print(f"   F1-score: {f1_score(y_train, y_pred_cv_extended):.3f}")

fig, axes = plt.subplots(1, 2, figsize=(12, 5))

axes[0].plot(recall_cv_base, precision_cv_base, 
            label=f'–ë–∞–∑–æ–≤–∞—è (AUC={pr_auc_cv_base:.3f})', 
            alpha=0.7, linewidth=2)
axes[0].plot(recall_cv_extended, precision_cv_extended, 
            label=f'–° –Ω–æ–≤—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏ (AUC={pr_auc_cv_extended:.3f})', 
            linewidth=2)
axes[0].set_xlabel('Recall')
axes[0].set_ylabel('Precision')
axes[0].set_title('PR-–∫—Ä–∏–≤—ã–µ –Ω–∞ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

metrics = ['Precision', 'Recall', 'F1', 'PR-AUC']
cv_base_scores = [
    precision_score(y_train, y_pred_cv_base),
    recall_score(y_train, y_pred_cv_base),
    f1_score(y_train, y_pred_cv_base),
    pr_auc_cv_base
]
cv_extended_scores = [
    precision_score(y_train, y_pred_cv_extended),
    recall_score(y_train, y_pred_cv_extended),
    f1_score(y_train, y_pred_cv_extended),
    pr_auc_cv_extended
]

x = np.arange(len(metrics))
width = 0.35
axes[1].bar(x - width/2, cv_base_scores, width, label='–ë–∞–∑–æ–≤–∞—è', alpha=0.7)
axes[1].bar(x + width/2, cv_extended_scores, width, label='–° –Ω–æ–≤—ã–º–∏ –ø—Ä–∏–∑–Ω–∞–∫–∞–º–∏', alpha=0.7)
axes[1].set_xlabel('–ú–µ—Ç—Ä–∏–∫–∞')
axes[1].set_ylabel('–ó–Ω–∞—á–µ–Ω–∏–µ')
axes[1].set_title('–°—Ä–∞–≤–Ω–µ–Ω–∏–µ –º–µ—Ç—Ä–∏–∫ –Ω–∞ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏')
axes[1].set_xticks(x)
axes[1].set_xticklabels(metrics, rotation=45)
axes[1].legend()
axes[1].set_ylim([0, 1])

plt.tight_layout()
plt.show()

print("\n" + "=" * 70)
print("–ò–¢–û–ì–ò –°–†–ê–í–ù–ï–ù–ò–Ø –ù–ê –ö–†–û–°–°-–í–ê–õ–ò–î–ê–¶–ò–ò")
print("=" * 70)

print(f"PR-AUC: {pr_auc_cv_extended - pr_auc_cv_base:+.4f}")
print(f"Precision: {precision_score(y_train, y_pred_cv_extended) - precision_score(y_train, y_pred_cv_base):+.3f}")
print(f"Recall: {recall_score(y_train, y_pred_cv_extended) - recall_score(y_train, y_pred_cv_base):+.3f}")
print(f"F1: {f1_score(y_train, y_pred_cv_extended) - f1_score(y_train, y_pred_cv_base):+.3f}")

print(f"–£–ª—É—á—à–µ–Ω–∏–µ PR-AUC: {pr_auc_cv_extended - pr_auc_cv_base:.4f}")
```

## –≠—Ç–∞–ø 6. –≠–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç—ã —Å –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏

1. –ü–µ—Ä–µ—á–∏—Å–ª–∏—Ç–µ –≤—Å–µ –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã, —Å –∫–æ—Ç–æ—Ä—ã–º–∏ –ø–ª–∞–Ω–∏—Ä—É–µ—Ç–µ —ç–∫—Å–ø–µ—Ä–∏–º–µ–Ω—Ç–∏—Ä–æ–≤–∞—Ç—å.

2. –ü—Ä–æ–≤–µ–¥–∏—Ç–µ —Å–∏—Å—Ç–µ–º–∞—Ç–∏—á–µ—Å–∫–∏–π –ø–µ—Ä–µ–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –¥–ª—è `LogisticRegression`, –≤—ã–ø–æ–ª–Ω–∏—Ç–µ –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—é –¥–ª—è –∫–∞–∂–¥–æ–π –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏–∏.

3. –°–æ—Å—Ç–∞–≤—å—Ç–µ —Ç–∞–±–ª–∏—Ü—É —Å —Ä–µ–∑—É–ª—å—Ç–∞—Ç–∞–º–∏.

4. –í—ã–±–µ—Ä–∏—Ç–µ –ª—É—á—à—É—é –º–æ–¥–µ–ª—å, –æ—Ä–∏–µ–Ω—Ç–∏—Ä—É—è—Å—å –Ω–∞ –∑–∞–¥–∞–Ω–Ω—É—é –º–µ—Ç—Ä–∏–∫—É –∫–∞—á–µ—Å—Ç–≤–∞.

```python
C_values = [0.01, 0.1, 1, 10, 100]
class_weight_options = [None, 'balanced', {0: 1, 1: 5}]

print("–ü–∞—Ä–∞–º–µ—Ç—Ä—ã:")
print(f"C (—Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏—è): {C_values}")
print(f"class_weight (–≤–µ—Å–∞ –∫–ª–∞—Å—Å–æ–≤): {[str(w) for w in class_weight_options]}")
print(f"–í—Å–µ–≥–æ –∫–æ–º–±–∏–Ω–∞—Ü–∏–π: {len(C_values) * len(class_weight_options)}")
```

<div class="alert alert-success">
<h2> –ö–æ–º–º–µ–Ω—Ç–∞—Ä–∏–π —Ä–µ–≤—å—é–µ—Ä–∞  <a class="tocSkip"> </h2>

üëç –û—Ç–ª–∏—á–Ω—ã–π –≤—ã–±–æ—Ä –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤, –¥–ª—è —Ä–µ–≥—É–ª—è—Ä–∏–∑–∞—Ü–∏–∏ –∏—Å–ø–æ–ª—å–∑—É–µ–º –ª–æ–≥–∞—Ä–∏—Ñ–º–∏—á–µ—Å–∫—É—é —à–∫–∞–ª—É.

–ü–æ–ª—É—á–∞–µ—Ç—Å—è –º–∞—Å—à—Ç–∞–±–Ω—ã–π –ø–µ—Ä–µ–±–æ—Ä.

```python
results_extended = []

for C_val in C_values:
    for class_w in class_weight_options:
        
        print(f"\nüîç –¢–µ—Å—Ç–∏—Ä—É–µ–º: C={C_val}, class_weight={class_w}")
        
        model = LogisticRegression(
            C=C_val,
            class_weight=class_w,
            random_state=RANDOM_STATE,
            max_iter=1000,
            solver='liblinear'
        )
        
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)
        
        y_pred_proba_cv = cross_val_predict(
            model, 
            X_train_processed_extended,
            y_train,
            cv=cv,
            method='predict_proba',
            n_jobs=-1
        )[:, 1]
        
        y_pred_cv = cross_val_predict(
            model,
            X_train_processed_extended,
            y_train,
            cv=cv,
            method='predict',
            n_jobs=-1
        )
        
        precision_cv, recall_cv, _ = precision_recall_curve(y_train, y_pred_proba_cv)
        pr_auc_cv = auc(recall_cv, precision_cv)
        
        precision_cv_score = precision_score(y_train, y_pred_cv)
        recall_cv_score = recall_score(y_train, y_pred_cv)
        f1_cv_score = f1_score(y_train, y_pred_cv)
        
        results_extended.append({
            'C': C_val,
            'class_weight': str(class_w),
            'pr_auc_cv': pr_auc_cv,
            'precision_cv': precision_cv_score,
            'recall_cv': recall_cv_score,
            'f1_cv': f1_cv_score
        })
        
        print(f"   PR-AUC (CV): {pr_auc_cv:.4f}")
        print(f"   Precision (CV): {precision_cv_score:.3f}")
        print(f"   Recall (CV): {recall_cv_score:.3f}")
        print(f"   F1 (CV): {f1_cv_score:.3f}")
```

–ò —Å–¥–µ–ª–∞–µ–º —Ç–∞–±–ª–∏—á–∫—É.

```python
results_extended_df = pd.DataFrame(results_extended)

results_extended_sorted = results_extended_df.sort_values('pr_auc_cv', ascending=False)

print("\n–†–µ–π—Ç–∏–Ω–≥ –º–æ–¥–µ–ª–µ–π:")
print("-" * 100)

for i, (_, row) in enumerate(results_extended_sorted.iterrows(), 1):
    print(f"{i:2}. C={row['C']:6} | class_weight={row['class_weight']:20} | ", end="")
    print(f"PR-AUC: {row['pr_auc_cv']:.4f} | ", end="")
    print(f"Precision: {row['precision_cv']:.3f} | ", end="")
    print(f"Recall: {row['recall_cv']:.3f} | ", end="")
    print(f"F1: {row['f1_cv']:.3f}")

best_extended_result = results_extended_sorted.iloc[0]
best_extended_C = best_extended_result['C']
best_extended_class_weight = best_extended_result['class_weight']
best_extended_pr_auc = best_extended_result['pr_auc_cv']

print(f"\n–õ—É—á—à–∞—è –º–æ–¥–µ–ª–∏ –ø–æ –∏—Ç–æ–≥–∞–º –ø–µ—Ä–µ–±–æ—Ä–∞:")
print(f"C = {best_extended_C}")
print(f"class_weight = {best_extended_class_weight}")
print(f"PR-AUC (–∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏—è) = {best_extended_pr_auc:.4f}")
```

–í –∏—Ç–æ–≥–µ —Å –ø–æ–º–æ—â—å—é –∏–∑–º–µ–Ω–µ–Ω–∏–π –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ –ø–æ–ª—É—á–∏–ª–æ—Å—å –ø–æ–ª—É—á–∏–ª–æ—Å—å —É–≤–µ–ª–∏—á–∏—Ç—å –∑–Ω–∞—á–µ–Ω–∏–µ –∫–ª—é—á–µ–≤–æ–π –º–µ—Ç—Ä–∏–∫–∏ PR-AUC —Å 0.6702 –¥–æ 0.6828, —Ç–æ –µ—Å—Ç—å +0.0126

–û–±—ä–µ–¥–∏–Ω–∏—Ç–µ –ª—É—á—à—É—é –∫–æ–Ω—Ñ–∏–≥—É—Ä–∞—Ü–∏—é –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤ —Å –æ–ø—Ç–∏–º–∞–ª—å–Ω—ã–º –Ω–∞–±–æ—Ä–æ–º –ø—Ä–∏–∑–Ω–∞–∫–æ–≤. –û–±—É—á–∏—Ç–µ –º–æ–¥–µ–ª—å –Ω–∞ –≤—Å–µ—Ö –¥–∞–Ω–Ω—ã—Ö –¥–ª—è –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ –∏ –ø—Ä–æ–≤–µ–¥–∏—Ç–µ —Ñ–∏–Ω–∞–ª—å–Ω—É—é –æ—Ü–µ–Ω–∫—É –Ω–∞ –æ—Ç–ª–æ–∂–µ–Ω–Ω–æ–π —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ.


```python
final_model = LogisticRegression(
    C=0.01,
    class_weight={0: 1, 1: 5},
    random_state=RANDOM_STATE,
    max_iter=1000,
    solver='liblinear'
)

final_model.fit(X_train_processed_extended, y_train)

X_test_extended_processed = full_preprocessor_extended.transform(X_test_extended)

y_pred_final = final_model.predict(X_test_extended_processed)
y_pred_proba_final = final_model.predict_proba(X_test_extended_processed)[:, 1]

precision_final, recall_final, _ = precision_recall_curve(y_test, y_pred_proba_final)
pr_auc_final = auc(recall_final, precision_final)

precision_score_final = precision_score(y_test, y_pred_final)
recall_score_final = recall_score(y_test, y_pred_final)
f1_score_final = f1_score(y_test, y_pred_final)
accuracy_final = accuracy_score(y_test, y_pred_final)

print("\n–†–µ–∑—É–ª—å—Ç–∞—Ç—ã –Ω–∞ —Ç–µ—Å—Ç–æ–≤–æ–π –≤—ã–±–æ—Ä–∫–µ:")
print(f"PR-AUC: {pr_auc_final:.4f}")
print(f"Precision (–∫–ª–∞—Å—Å 1): {precision_score_final:.3f}")
print(f"Recall (–∫–ª–∞—Å—Å 1): {recall_score_final:.3f}")
print(f"F1-score (–∫–ª–∞—Å—Å 1): {f1_score_final:.3f}")
print(f"Accuracy: {accuracy_final:.3f}")

print(f"\nClassification Report:")
print(classification_report(y_test, y_pred_final))
```
